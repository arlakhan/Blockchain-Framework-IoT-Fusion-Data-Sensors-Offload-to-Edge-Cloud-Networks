```python
from flask import Flask, request, jsonify, render_template_string
from flask_ngrok import run_with_ngrok
import hashlib
import time

app = Flask(__name__)
run_with_ngrok(app)


# Blockchain class to manage the chain
class Blockchain:
    def __init__(self):
        self.chain = []
        self.difficulty = 2
        self.create_genesis_block()

    def create_genesis_block(self):
        genesis_block = Block(0, "0", time.time(), "Genesis Block", "0")
        genesis_block.hash = genesis_block.calculate_hash()
        self.chain.append(genesis_block)

    def add_block(self, data):
        previous_block = self.chain[-1]
        new_block = Block(len(self.chain), previous_block.hash, time.time(), data, "")
        self.mine_block(new_block)
        self.chain.append(new_block)

    def mine_block(self, block):
        while block.hash[:self.difficulty] != "0" * self.difficulty:
            block.nonce += 1
            block.hash = block.calculate_hash()

    def get_chain(self):
        return [block.to_dict() for block in self.chain]


# Block class for individual blocks
class Block:
    def __init__(self, index, previous_hash, timestamp, data, hash):
        self.index = index
        self.previous_hash = previous_hash
        self.timestamp = timestamp
        self.data = data
        self.hash = hash
        self.nonce = 0

    def calculate_hash(self):
        block_string = f"{self.index}{self.previous_hash}{self.timestamp}{self.data}{self.nonce}"
        return hashlib.sha256(block_string.encode("utf-8")).hexdigest()

    def to_dict(self):
        return {
            "index": self.index,
            "previous_hash": self.previous_hash,
            "timestamp": self.timestamp,
            "data": self.data,
            "hash": self.hash,
            "nonce": self.nonce,
        }


# Hospital Nodes
hospitals = {
    "Hospital_A": Blockchain(),
    "Hospital_B": Blockchain(),
}

# HTML templates
HOME_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Hospital Blockchain</title>
</head>
<body>
    <h1>Hospital Blockchain System</h1>
    <h3>Select Hospital</h3>
    <form action="/hospital" method="get">
        <label for="hospital">Choose a hospital:</label>
        <select name="hospital" id="hospital">
            <option value="Hospital_A">Hospital A</option>
            <option value="Hospital_B">Hospital B</option>
        </select>
        <button type="submit">Go</button>
    </form>
</body>
</html>
"""

HOSPITAL_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{{ hospital_name }}</title>
</head>
<body>
    <h1>{{ hospital_name }}</h1>
    <form action="/add_patient" method="post">
        <h3>Add Patient Data</h3>
        <input type="hidden" name="hospital" value="{{ hospital_name }}">
        <label for="patient_name">Name:</label>
        <input type="text" id="patient_name" name="patient_name" required><br><br>
        <label for="patient_age">Age:</label>
        <input type="number" id="patient_age" name="patient_age" required><br><br>
        <label for="condition">Condition:</label>
        <input type="text" id="condition" name="condition" required><br><br>
        <button type="submit">Add Patient</button>
    </form>
    <hr>
    <form action="/share_data" method="post">
        <h3>Share Patient Data</h3>
        <input type="hidden" name="source_hospital" value="{{ hospital_name }}">
        <label for="patient_name_share">Patient Name:</label>
        <input type="text" id="patient_name_share" name="patient_name" required><br><br>
        <label for="target_hospital">Target Hospital:</label>
        <select name="target_hospital" id="target_hospital">
            {% for hospital in hospitals %}
            {% if hospital != hospital_name %}
            <option value="{{ hospital }}">{{ hospital }}</option>
            {% endif %}
            {% endfor %}
        </select><br><br>
        <button type="submit">Share Data</button>
    </form>
    <hr>
    <form action="/view_blockchain" method="get">
        <input type="hidden" name="hospital" value="{{ hospital_name }}">
        <button type="submit">View Blockchain</button>
    </form>
    <hr>
    <form action="/" method="get">
        <button type="submit">Back to Home</button>
    </form>
</body>
</html>
"""

BLOCKCHAIN_TEMPLATE = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>{{ hospital_name }} Blockchain</title>
</head>
<body>
    <h1>{{ hospital_name }} Blockchain Data</h1>
    <ul>
        {% for block in blockchain %}
        <li>
            <strong>Block #{{ block.index }}</strong><br>
            Previous Hash: {{ block.previous_hash }}<br>
            Timestamp: {{ block.timestamp }}<br>
            Data: {{ block.data }}<br>
            Hash: {{ block.hash }}<br>
            Nonce: {{ block.nonce }}
        </li>
        <hr>
        {% endfor %}
    </ul>
    <form action="/" method="get">
        <button type="submit">Back to Home</button>
    </form>
</body>
</html>
"""


@app.route("/")
def home():
    """Render the home page."""
    return render_template_string(HOME_TEMPLATE)


@app.route("/hospital")
def hospital_page():
    """Render the hospital-specific page."""
    hospital_name = request.args.get("hospital")
    return render_template_string(
        HOSPITAL_TEMPLATE, hospital_name=hospital_name, hospitals=hospitals.keys()
    )


@app.route("/add_patient", methods=["POST"])
def add_patient():
    """Add a patient's data to the selected hospital's blockchain."""
    hospital_name = request.form["hospital"]
    patient_name = request.form["patient_name"]
    patient_age = request.form["patient_age"]
    condition = request.form["condition"]

    # Create patient data
    patient_data = {
        "type": "Patient",
        "name": patient_name,
        "age": int(patient_age),
        "condition": condition,
    }

    # Add to blockchain
    hospitals[hospital_name].add_block(patient_data)
    return f"<h1>Patient added to {hospital_name} successfully!</h1><form action='/'><button type='submit'>Back to Home</button></form>"


@app.route("/share_data", methods=["POST"])
def share_data():
    """Share patient data between hospitals."""
    source_hospital = request.form["source_hospital"]
    target_hospital = request.form["target_hospital"]
    patient_name = request.form["patient_name"]

    # Find the patient data in the source hospital's blockchain
    for block in hospitals[source_hospital].chain:
        if block.data.get("type") == "Patient" and block.data.get("name") == patient_name:
            # Share the data
            hospitals[target_hospital].add_block(block.data)
            return f"<h1>Patient data shared from {source_hospital} to {target_hospital} successfully!</h1><form action='/'><button type='submit'>Back to Home</button></form>"

    return f"<h1>Patient not found in {source_hospital}!</h1><form action='/'><button type='submit'>Back to Home</button></form>"


@app.route("/view_blockchain", methods=["GET"])
def view_blockchain():
    """View the blockchain of a specific hospital."""
    hospital_name = request.args.get("hospital")
    blockchain_data = hospitals[hospital_name].get_chain()
    return render_template_string(
        BLOCKCHAIN_TEMPLATE, hospital_name=hospital_name, blockchain=blockchain_data
    )


if __name__ == "__main__":
    app.run()
```
